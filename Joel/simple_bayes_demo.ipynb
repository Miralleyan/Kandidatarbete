```python
import numpy as np
import matplotlib.pyplot as plt
```

$p(μ | x) = \frac{p(x | μ)p(μ)}{p(x)}$
in bayesian statistics the parameters are not numbers, but statistical distributions.

* We have a hypothesis for the distribution of the parameters (so called prior for parameters)
* We want to find the distribution of the parameter given some data that we've observed -- the so called posterior distribution of the parameter
* $p(x)$ will just normalize the result, so if we normalize μ | x in another way, we don't need to hypothesise about the distribution of x.

```python
μs = np.linspace(-5, 5, 100) # assume that μ is roughly contained in the range (-5, 5)
N = 20
x = np.random.randn(N) + 2
```

hypothesis / prior for mu is normally distributed (variance 1)

```python
prior = np.exp(-0.5*(μs)**2)
prior /= np.sum(prior) # normalize so that the distribution of the parameter have sum 1
```

```python
plt.plot(μs, prior)
```

Demonstrate algebra for posterior on whiteboard!

```python
def f(μ):
    log_posterior_unnormalized = -0.5*np.sum((x-μ)**2) - 0.5*μ**2
    return np.exp(log_posterior_unnormalized)
f = np.vectorize(f)
```

```python
posterior = f(μs)
posterior /= np.sum(posterior)
plt.plot(μs, posterior)
```
