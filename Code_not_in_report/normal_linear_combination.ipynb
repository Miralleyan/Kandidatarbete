```python
import torch
import linear_combination_optimizer as lco
torch.manual_seed(1)

N = 200
x = torch.linspace(-4,4, N)
y = (torch.randn(N)-4) * (x**2) + (-2+torch.randn(N)) * x + (3+torch.randn(N))

opt = lco.Optimizer(x, y, order=3)
opt.optimize(epochs = 300, lr=0.1)
opt.visualize()
```
