```python
import torch
import pytorch_measure as pm
import matplotlib.pylab as plt
import numpy as np

def test_sample_method():
    a=torch.tensor([0.1,0.1,0.3,0.1,0.4])
    b=torch.tensor([1.,2.,3.,4.,5.])

    c=pm.Pytorch_measure(b,a)

    size = 5000
    sample = c.sample(size)
    sample_frequencies = [0]*len(b)
    mapping = {elem.item():i for i, elem in enumerate(b)}
    for elem in sample:
        sample_frequencies[mapping[elem.item()]] += 1
    generated = []
    for i, elem in enumerate(sample_frequencies):
        generated.append(elem/size)
        plt.bar(i+1, elem)

    expected = [round(elem.item(),1) for elem in a]
    print("Frequencies:",sample_frequencies)
    print("Expected:", expected)
    print("Generated:", generated)
    print("Difference:", [round(abs(exp - gen),5) for exp,gen in zip(expected, generated)])
    plt.show()

t = torch.tensor([0,1,2])

def get_item(tensor_single):
    return tensor_single.item()
    
for elem in map(get_item, t):
    print(elem)

test_sample_method()

```
